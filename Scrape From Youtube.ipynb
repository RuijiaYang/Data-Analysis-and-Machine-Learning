## Use the official API to Scrape Data

We again use several accounts to conduct the scraping at the same time. 

```python
api_key="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
```

```python
parameters_1 = {"part":"snippet",
             "maxResults":10,
             "order":"date",
             "pageToken":"",
              "q":"",
             "key":api_key,
             "type":"video",}
url_1 = "https://www.googleapis.com/youtube/v3/search"
```

```python
for t in range(0,500):
    parameters_1["q"] = movie_title[t]
    page = requests.request(method="get",url=url_1,params=parameters_1)
    j_results = json.loads(page.text)
    data = j_results['items']
    for i in data:
        #print(i['kind'],'+',i['snippet']['title'],'+',i['id']['videoId'],'+',i['snippet']['channelId'],'+',i['snippet']['publishedAt'])
        parameters_2={"part":"statistics,snippet",
                      "maxResults":10,
                      "id":"",
                      "key":api_key,
                     }
    
        url_2 = "https://www.googleapis.com/youtube/v3/videos"
        parameters_2["id"] = i['id']['videoId']
        page_2 = requests.request(method="get",url=url_2,params=parameters_2)
        j_results_2 = json.loads(page_2.text)
        data_2 = j_results_2['items']
        for d in data_2:
            try:
                print(movie_title[t],'=',i['kind'],'=',i['snippet']['title'],'=',i['snippet']['description'],'=',i['id']['videoId'],'=',i['snippet']['channelId'],'=',i['snippet']['publishedAt'],'=',d['snippet']['tags'],'=',d['snippet']['channelTitle'],'=',d['statistics']['viewCount'],'=',d['statistics']['likeCount'])
            except:
                print("")
```
